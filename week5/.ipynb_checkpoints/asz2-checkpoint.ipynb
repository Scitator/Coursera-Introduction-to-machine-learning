```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import log_loss
import matplotlib.pyplot as plt
%matplotlib inline
import operator
```

```python
df = pd.read_csv('gbm-data.csv', header=0)
```

```python
y = np.array(df.values[:, 0])
X = np.array(df.values[:, 1:])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=241)
```

```python
for learning_rate in [1, 0.5, 0.3, 0.2, 0.1]:
    gbc = GradientBoostingClassifier(n_estimators=250, verbose=True, random_state=241, learning_rate=learning_rate)
    gbc.fit(X_train, y_train)
    test_loss = []
    for y_pred in gbc.staged_predict_proba(X_test):
        test_loss.append(log_loss(y_test, y_pred))
    train_loss = []
    for y_pred in gbc.staged_predict_proba(X_train):
        train_loss.append(log_loss(y_train, y_pred))
    plt.figure()
    plt.plot(test_loss , 'r', linewidth=2)
    plt.plot(train_loss, 'g', linewidth=2)
    plt.legend(['test', 'train'])
```

```python
for learning_rate in [0.2]:
    gbc = GradientBoostingClassifier(n_estimators=250, verbose=True, random_state=241, learning_rate=learning_rate)
    gbc.fit(X_train, y_train)
    test_loss = {}
    step = 0
    for y_pred in gbc.staged_predict_proba(X_test):
        step = step + 1
        test_loss[step] = log_loss(y_test, y_pred)
    test_loss_sorted = sorted(test_loss.items(), key=operator.itemgetter(1), reverse=True)
    for x in test_loss_sorted:
        print x
```

```python
for learning_rate in [0.2]:
    gbc = GradientBoostingClassifier(n_estimators=37, verbose=True, random_state=241, learning_rate=learning_rate)
    gbc.fit(X_train, y_train)
    y_pred = gbc.predict_proba(X_test):
    test_loss = log_loss(y_test, y_pred)
    print test_loss
```
